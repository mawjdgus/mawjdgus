```python
class Simple1D(nn.Module):
    def __init__(self):
        super(Simple1D, self).__init__()
        self.cnn1 = nn.Conv1d(1,10, 3, 1,padding=1)
        self.batch = nn.BatchNorm1d(10)
        self.act2 = nn.ReLU()
        self.drop1 = nn.Dropout(0.2)

        self.cnn2 = nn.Conv1d(10,64, 3, 1,padding=1)
        self.batch2 = nn.BatchNorm1d(64)
        self.act3 = nn.ReLU()
        self.drop2 = nn.Dropout(0.2)


        self.cnn3 = nn.Conv1d(64,128, 3, 1,padding=1)
        self.batch3 = nn.BatchNorm1d(128)
        self.act4 = nn.ReLU()
        self.drop3 = nn.Dropout(0.2)


        self.cnn4 = nn.Conv1d(128,256, 3, 1,padding=1)
        self.batch4 = nn.BatchNorm1d(256)
        self.act5 = nn.ReLU()
        self.drop4 = nn.Dropout(0.2)

        self.cnn5 = nn.Conv1d(256,1, 3, 1,padding=1)
        self.batch5 = nn.BatchNorm1d(1)
        self.act6 = nn.ReLU()
        self.drop5 = nn.Dropout(0.2)

        self.li3 = nn.Linear(357,512)
        self.act7 = nn.ReLU()
        self.drop6 = nn.Dropout(0.2)

        self.li4 = nn.Linear(512,1024)
        self.act8 = nn.ReLU()
        self.drop7 = nn.Dropout(0.2)
        
        self.li5 = nn.Linear(1024,512)
        self.act9 = nn.ReLU()
        self.drop8 = nn.Dropout(0.2)
        
        self.li6 = nn.Linear(512,31)

        


    def forward(self, x):

        
        x = self.cnn1(x)
        x = self.batch(x)
        x = self.act2(x)
        x = self.drop1(x)
        x = self.cnn2(x)
        x = self.batch2(x)
        x = self.act3(x)
        x = self.drop2(x)
        x = self.cnn3(x)
        x = self.batch3(x)
        x = self.act4(x)
        x = self.drop3(x)
        x = self.cnn4(x)
        x = self.batch4(x)
        x = self.act5(x)
        x = self.drop4(x)
        x = self.cnn5(x)
        x = self.batch5(x)
        x = self.act6(x)
        x = self.drop5(x)
        

        x = self.li3(x)
        x = self.act7(x)
        x = self.drop6(x)

        x = self.li4(x)
        x = self.act8(x)
        x = self.drop7(x)

        x = self.li5(x)
        x = self.act9(x)
        x = self.drop8(x)



        x = self.li6(x)
        
        
        
        
        

        return x
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
stratifiedKFold = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
y_label = []
for index in df_drop_duplicate.index:
    _, y , _ = map(int,index.split(','))
    y_label.append(y)
```

```python
class TrainDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe


    def __getitem__(self, index):

        rssi = self.dataframe[index][0]
        label = self.dataframe[index][1]

        
        return rssi , label

    def __len__(self):
        return len(self.dataframe)

```

```python

model = Simple1D()
model.to(device,dtype=torch.float64)

```

```python
num_epochs = 50
for fold, (train_ids, val_ids) in enumerate(stratifiedKFold.split(b,y_label)):
    torch.cuda.empty_cache()
    model = Simple1D()
    model.to(device,dtype=torch.float64)

    train_data = [b[ids] for ids in train_ids]
    val_data = [b[ids] for ids in val_ids]
    
    train_dataset = TrainDataset(train_data)
    val_dataset = TrainDataset(val_data)

    print(f'FOLD {fold}')
    print('-------------------------')

    train_loader = DataLoader(train_dataset, batch_size=50, drop_last=True)
    val_loader = DataLoader(val_dataset, batch_size=50, drop_last=True)

    
    criterion = nn.CrossEntropyLoss()
    opt = optim.Adam(model.parameters(), lr=0.0002)

    for epoch in range(num_epochs):
        print(f'=======EPOCH : {epoch}=======')
        model.train()
        train_loss = 0
        train_acc = 0
        val_loss = 0
        val_acc = 0
        epoch_loss = 0
        epoch_acc = 0
        epoch_val_loss = 0
        epoch_val_acc = 0

        for i, data in enumerate(tqdm(train_loader, 0)):
            inputs, targets = data
            inputs = inputs.unsqueeze(1)
            targets = targets-1
            inputs, targets = inputs.to(device), targets.to(device)
            

            opt.zero_grad()
            
            pred = model(inputs)
            pred = pred.squeeze(1)

            loss = criterion(pred, targets)
            _,preds = torch.max(pred, 1)

            loss.backward()
            opt.step()
            
            train_loss += loss.item()
            train_acc += torch.sum(preds==targets.data)
        print(len(train_loader.dataset))
        epoch_loss = train_loss /len(train_loader.dataset)
        epoch_acc = train_acc / len(train_loader.dataset)
        print(epoch_loss)
        print(epoch_acc*100)

        with torch.no_grad():
            # Iterate over the test data and generate predictions
            for i, data in enumerate(tqdm(val_loader)):
                model.eval()
                inputs, targets = data
                targets = targets-1
                inputs= inputs.unsqueeze(1)
                inputs, targets = inputs.to(device), targets.to(device)


                pred = model(inputs)
                pred = pred.squeeze(1)
                loss = criterion(pred, targets)
                # Set total and correct
                _, preds= torch.max(pred, 1)
                val_loss += loss.item()
                val_acc += torch.sum(preds == targets.data)
            epoch_val_loss = val_loss / len(val_loader.dataset)
            print(len(val_loader.dataset))
            epoch_val_acc = val_acc / len(val_loader.dataset)
            print(epoch_val_loss)
            print(epoch_val_acc)
        print('Accuracy for fold %d: %d %%' % (fold, 100.0 * epoch_val_acc))
        print('--------------------------------')
```
